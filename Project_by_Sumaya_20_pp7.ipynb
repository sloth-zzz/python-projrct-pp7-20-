```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("environmental_data.csv")

# Plotting the data
plt.figure(figsize=(8, 6))  # Set the figure size

# Plotting each column against the index
plt.plot(df.index, df['Temperature'], label='Temperature', marker='o',ls="dotted")
plt.plot(df.index, df['Humidity'], label='Humidity', marker='*',ls="dashed")
plt.plot(df.index, df['Pressure'], label='Pressure', marker='^',ls="dashdot")

# Addition of titles, labels, and legend
plt.title('Line Graph for Temperature, Humidity, and Pressure from csv file called environmental_data')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.grid(color="green",ls="--",linewidth="0.5")

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Data
data = {
    "Temperature": [23, 39, 8, 46, 10, 39, 29, 46, -2, 35, 34, 14, 21, 21, 14, 37, 45, 18, 12, 13],
    "Humidity": [63, 87, 8, 50, 100, 97, 21, 18, 28, 44, 80, 29, 67, 8, 14, 77, 41, 60, 42, 92],
    "Pressure": [947, 1073, 941, 930, 1083, 991, 1074, 1018, 906, 982, 1068, 1005, 1007, 906, 910, 958, 945, 995, 1064, 1027],
}

# Creating DataFrame
df = pd.DataFrame(data)

# Set the style
sns.set_style("whitegrid")

# Creating a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x=df.index, y="Temperature", color="blue", label="Temperature")
sns.barplot(data=df, x=df.index, y="Humidity", color="green", label="Humidity", alpha=0.7)
sns.barplot(data=df, x=df.index, y="Pressure", color="red", label="Pressure", alpha=0.4)

# title of the graph
plt.title("Bar Graph of Temperature, Humidity, and Pressure", fontsize=16)
plt.xlabel("Index", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import cumsum
from bokeh.palettes import Category20c
from bokeh.models import HoverTool
import numpy as np
output_notebook()

df = pd.read_csv('environmental_data.csv')

# Aggregate data for a specific metric
# Example: Summing up Temperature, Humidity, and Pressure for a pie chart
data = df[['Temperature', 'Humidity', 'Pressure']].sum().reset_index()
data.columns = ['Metric', 'Value']

# Preparing data for the pie chart
data['Angle'] = data['Value'] / data['Value'].sum() * 2 * np.pi  # Calculate angles for the pie chart
data['Color'] = Category20c[len(data)]  # Assign colors to each slice

#  pie chart creation
p = figure(height=350, title="Pie Chart for Metrics fron environmental_data of Temperature,Humidity & Pressure", toolbar_location=None, tools="hover", tooltips="@Metric: @Value", x_range=(-0.5, 1.0))

# Addition of the wedges for the pie chart
p.wedge(
    x=0, y=1, radius=0.4,
    start_angle=cumsum('Angle', include_zero=True),
    end_angle=cumsum('Angle'),
    line_color="white", fill_color='Color', legend_field='Metric',
    source=data
)

# hover tool use
hover = HoverTool()
hover.tooltips = [("Metric", "@Metric"), ("Value", "@Value")]
p.add_tools(hover)


p.axis.axis_label = None
p.axis.visible = False
p.grid.grid_line_color = None

# Show the pie chart
show(p)
```
